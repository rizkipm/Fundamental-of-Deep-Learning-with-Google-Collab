#Runtime


* Runtime type
  * Python 3 : Allows you to change programming language in collab
* Hardware accelerator
  * CPU: Simple computations
  * A100 GPU: Powerful GPU suitable for **High performance** computing, machine learning and DL task.
  * L4 GPU: Good performance for inference and some tpyes of training
  * T4 GPU (**COMMON**): less powerful than A100, but capable to handle machine learning and DL
  * TPU V2-8: specifically for TensorFlow-based DL
  * High-RAM: Access more memory, useful for memory-intensive tasks like processing large datasets or running large models


How to Choose?

1.	If you are working on **small** or **basic** tasks, a CPU should be enough.
1.	For tasks that involve moderate machine learning or small neural network training, a T4 GPU is typically a good balance between performance and availability.
1.	If you are working on training larger models or need faster performance, you might want to go for an A100 GPU.
1.	TPUs are best for TensorFlow-based models and can significantly speed up training.

Tips; Always stick to T4 unless you want more GPUs


1. Start with a T4: Begin by running your workload on a T4 and monitor its performance. Track key metrics like runtime, memory utilization, and batch size limitations.

1. Analyze Performance: Look at GPU utilization. If it’s consistently hitting limits (e.g., memory or processing power), and processing times are long, it’s an indicator that your workload demands more power.

1. Switch to an A100 if Needed: If you see substantial runtime improvements or memory benefits by upgrading to an A100, make the switch. This approach minimizes unnecessary costs while still delivering good performance when required.

#Built in library


#### **Core Python Libraries:**
- **os**: Access to operating system functionalities.
- **sys**: System-specific parameters and functions.
- **math**: Mathematical functions like logarithms, trigonometry, etc.
- **random**: Functions to generate random numbers and choices.
- **time**: Time-related functions.
- **datetime**: Date and time manipulation.
- **collections**: Specialized container datatypes like `namedtuple`, `deque`, etc.
- **itertools**: Functions to create iterators for efficient looping.
- **functools**: Higher-order functions like `reduce`, `partial`, etc.
- **re**: Regular expressions handling.

#### **Data Manipulation and Analysis:**
- **numpy**: Array manipulation and mathematical functions.
- **pandas**: Data manipulation and analysis.
- **scipy**: Scientific and technical computing.
- **matplotlib**: Plotting and visualization.
- **seaborn**: Statistical data visualization.
- **statsmodels**: Statistical modeling and econometrics.
- **sklearn** (scikit-learn): Machine learning library.
- **sympy**: Symbolic mathematics.

#### **Machine Learning and Deep Learning:**
- **tensorflow**: Deep learning framework (including Keras).
- **torch** (PyTorch): Deep learning framework.
- **keras**: High-level neural networks API.
- **xgboost**: Optimized gradient boosting.
- **lightgbm**: Gradient boosting framework.
- **catboost**: Gradient boosting library.

#### **Data Visualization:**
- **matplotlib**: Comprehensive library for static, animated, and interactive plots.
- **seaborn**: Statistical visualization based on matplotlib.
- **plotly**: Interactive plots and dashboards.
- **bokeh**: Interactive visualization library.

#### **Data Processing and I/O:**
- **requests**: HTTP library to send requests.
- **urllib**: URL handling and web interactions.
- **csv**: Reading and writing CSV files.
- **json**: JSON parsing and manipulation.
- **zipfile**: Handling ZIP file compression.
- **tarfile**: Handling TAR file compression.
- **io**: Core tools for working with streams.
- **PIL (Pillow)**: Image processing.

#### **Natural Language Processing (NLP):**
- **nltk**: Natural Language Toolkit.
- **spacy**: Industrial-strength NLP.
- **transformers** (Hugging Face): Transformers library for NLP models.

#### **Others for Deep Learning and Computer Vision:**
- **opencv**: OpenCV for computer vision tasks.
- **PIL**: Image processing (also accessible via the **Pillow** package).
- **imageio**: Reading and writing images.
- **tqdm**: Progress bar for loops.
- **h5py**: HDF5 file handling.
- **pickle**: Serializing and deserializing Python objects.

#### **Google Colab-Specific Libraries:**
- **google.colab**: Provides features such as file uploads, downloads, and integration with Google Drive.
- **IPython.display**: Display images, HTML, and LaTeX outputs within Colab.
